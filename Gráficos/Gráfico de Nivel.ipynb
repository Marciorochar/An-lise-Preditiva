```python
# Instalar bibliotecas
!pip install pandas matplotlib openpyxl

# Imports
import pandas as pd
import matplotlib.pyplot as plt

# Upload do arquivo Excel
from google.colab import files
uploaded = files.upload()

# Leitura do arquivo Excel
nome_arquivo = list(uploaded.keys())[0]
df = pd.read_excel(nome_arquivo, engine='openpyxl')

# Converter colunas
df['Timestamp'] = pd.to_datetime(df['Timestamp'], dayfirst=True, errors='coerce')
df['Camisetas_bÃ¡sicas_masculinas'] = pd.to_numeric(df['Camisetas_bÃ¡sicas_masculinas'], errors='coerce')
df = df.dropna(subset=['Timestamp', 'Camisetas_bÃ¡sicas_masculinas'])

# Ordenar por data
df = df.sort_values('Timestamp')

# Calcular o nÃ­vel mÃ©dio
nivel_medio = df['Camisetas_bÃ¡sicas_masculinas'].mean()

# GrÃ¡fico
plt.figure(figsize=(12, 6))
plt.plot(df['Timestamp'], df['Camisetas_bÃ¡sicas_masculinas'], label='Vendas DiÃ¡rias', color='steelblue')
plt.axhline(y=nivel_medio, color='red', linestyle='--', linewidth=2, label=f'NÃ­vel MÃ©dio: {nivel_medio:.2f}')

plt.title('GrÃ¡fico de NÃ­vel - Camisetas BÃ¡sicas Masculinas')
plt.xlabel('Data')
plt.ylabel('Quantidade de Vendas')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
